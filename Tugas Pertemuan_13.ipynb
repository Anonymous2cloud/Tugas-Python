```
import numpy as np
import matplotlib.pyplot as plt
```

# KURVA 1


```
def cos(frek,tipe,ht,hr,d):
  #frek = frekuensi (MHz)
  #tipe = urban, suburban, open
  #ht = tinggi tower (meter)
  #hr = tinggi user (meter)
  #d = jarak (meter)
    if frek in range(150,1501):
        Ca=69.55
        Cb=26.16
    elif frek in range(1501,2001):
        Ca=46.3
        Cb=33.9
    
    if tipe=='urban' :
        Cm=0
    elif tipe=='suburban' :
        Cm= -2*((np.log10((frek/28)**2))-5.4)
    elif tipe=='open' :
        Cm= -4.78*(np.log10(frek))+18.33*(np.log10(frek))-40.94
    else :
        Cm='tipe salah'
    
    ahr = (1.1*np.log10(frek)-0.7)*hr-(1.56*np.log10(frek)-1.8)
    Lp = Ca+(Cb*np.log10(frek))-(13.83*np.log(ht))-ahr+((44.9-6.55*np.log10(ht))*np.log10(d))-Cm
    
    return Lp
```

```
f=1800 #MHz
loc='urban'
loc2='open'
loc3='sub_urban'
tower=80 #m
user=1.5 #m
dmin=200 #m
dmax=700 #m
N=200 #titik
d=np.random.randint(dmin,dmax,N)
d2=np.random.randint(dmin,dmax,N)
d3=np.random.randint(dmin,dmax,N)
```

```
print(d,d2,d3)
```

```
LpUrban=cos(f,loc,tower,user,d)
LpOpen=cos(f,loc,tower,user,d2)
LpSub_urban=cos(f,loc,tower,user,d3)
print(LpUrban,LpOpen,LpSub_urban)
```

```
d_sort=d[np.argsort(d)]
d2_sort=d[np.argsort(d)]
d3_sort=d[np.argsort(d)]
Urban_sort=LpUrban[np.argsort(d)]
Open_sort=LpOpen[np.argsort(d2)]
Sub_urban_sort=LpSub_urban[np.argsort(d3)]
```

```
plt.figure()
plt.plot(d_sort,Urban_sort,label='Urban')
plt.plot(d2_sort,Open_sort,label='Open')
plt.plot(d3_sort,Sub_urban_sort,label='Sub Urban')
plt.legend()
plt.xlabel('Jarak (m)')
plt.ylabel('Path Loss (dB)')
plt.show()
```
